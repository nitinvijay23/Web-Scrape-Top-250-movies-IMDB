```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
#Get the response using the requests library
response = requests.get("http://www.imdb.com/chart/top")
response
```

```python
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
#Prettify the soup
print(soup.prettify())
```

```python
#Get the movies
#All the movies have the same td class titleColumn
movies_list = []
movies_class = soup.find_all('td', class_='titleColumn')
for movie in movies_class:
    moviename = movie.find('a').get_text()
    movies_list.append(moviename)
movies_list
```

```python
#Get the Release Years
Years = []
for movie in movies_class:
    year = movie.find('span').get_text()
    Years.append(year)
Years
```

```python
Updated_Years = []
for year in Years:
    y = year[1:5]
    y = int(y)
    Updated_Years.append(y)

Updated_Years
```

```python
ratings = soup.find_all('td', class_ = 'ratingColumn imdbRating')
ratings
```

```python
RatingsIMDB = []

for rating in ratings:
    r = rating.find('strong').get_text()
    r = float(r)
    RatingsIMDB.append(r)
```

```python
import pandas as pd
topmovies = pd.DataFrame({
        'moviename': movies_list,
        'Year of Released': Updated_Years,
        'Rating': RatingsIMDB
    })
topmovies
```
